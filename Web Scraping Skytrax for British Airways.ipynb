```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# loading the website in our variable
html_text = requests.get('https://www.airlinequality.com/airline-reviews/british-airways/').text
print(html_text)
```

```python
# parsing the website with lxml parser for scraping convinently
soup = BeautifulSoup(html_text, 'lxml')
```

```python
# list to store the reviews
revs_list = []
```

```python
i = 1
page_size = 100
for i in range(1, 10):
    print(f"Scraping the Reviews from Page : {i} ...")
    # updating the url to move on to the next page
    html_text = requests.get(f'https://www.airlinequality.com/airline-reviews/british-airways/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}').text

    soup = BeautifulSoup(html_text, 'lxml')
    
    # loading all the reviews 
    reviews = soup.find_all('div', class_ = "text_content")

    for review in reviews:
        revs_list.append(review.text)
```

```python
df = pd.DataFrame(revs_list, columns=['Reviews'])
```
